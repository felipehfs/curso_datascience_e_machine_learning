```python
import pandas as pd
import numpy as np
# DataSet
dados = {
    'nomes':['João', 'Felipe', 'Pedro', 'Judas', 'Maria', 'Matheus', 'Paulo', 'Enoque'],
    'sexo':['M', 'M', 'M', 'M', np.nan, 'F', np.nan, 'M'],
    'idade': [32, 14, 14, np.nan, np.nan, 56, 43, 28],
    'nota': [14, 11, 5, 32, np.nan, 32, 14, 15]
}

df = pd.DataFrame(dados)
```

```python
df
```

```python
# Removendo apenas as linhas que tem um NaN
df.dropna()
```

```python
# Remover as linhas que todas as colunas são NaN.
df.dropna(how='all') 
```

```python
# Remove linhas com 2 colunas NaN
df.dropna(thresh=2)
```

```python
# Preenche NaN com um valor default
df['sexo'].fillna('Não declarado', inplace=True)
```
