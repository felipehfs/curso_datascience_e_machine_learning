```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
salario = np.array([200, 300, 400, 500, 200])
plt.plot(salario, c='blue', ls='--', marker='s', ms=18)
plt.show()
```

```python
# Adicionando Label e legenda
salario_felipe = np.array([100, 340, 120, 400])
salario_vanessa = np.array([899, 840, 983, 1000])
plt.plot(salario_felipe, c='Black', ls='--', marker='s', ms=8, label='Salário do Felipe')
plt.plot(salario_vanessa, c='Red', ls='--', marker='^', ms=8, label='Salário da Vanessa')
plt.legend()
plt.show()
```

```python
salario_pedro = np.array([120, 145, 320, 100])
salario_jose = np.array([389, 123, 432, 289])
salario_guilherme = np.array([300, 492, 720])

plt.plot(salario_pedro, c='Blue', ls='--', marker='o', ms=8, label='Salário do Pedro')
plt.plot(salario_jose, c='Red', ls='--', marker='^', ms=8, label='Salário do José')
plt.plot(salario_guilherme, c='Black', ls='--', marker='s', label="Salário do Guilherme")
plt.legend(loc='upper left')
plt.show()
```
