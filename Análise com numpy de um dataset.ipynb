```python
import numpy as np

data = np.genfromtxt('iris.data', delimiter=',', usecols=(0, 1, 2, 3))
data
```

```python
data[:50, 0]
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(data[:50, 0], c='red', ls=':', marker='s', ms=8)
```

```python
plt.plot(data[:50, 0], label='Comp. da petals Iris-setosa', c='Red', ls=':', marker='s', ms=8)
plt.plot(data[50:100, 0], label='Comp. da petals Iris-versicolor', c='Black', ls=':', marker='o', ms=8)

plt.legend()
plt.show()
```
